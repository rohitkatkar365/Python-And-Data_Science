```python
import pandas as pd
```

```python
df = pd.read_csv("weather.csv")
df
```

```python
# Pivot allow you to reshape or transfrom data
df.pivot(index="date",columns="city")
```

```python
df.pivot(index="date",columns="city",values="humidity")
```

```python
# Pivot table is used to summerize and aggregrate data inside dataframe
df.pivot_table(index="city",columns="date")
```

```python
df.pivot_table(index="city",columns="date",margins=True)
```
