```python
# TimeSeries is a set of data points indexed in time order
import pandas as pd
df = pd.read_csv("aapl.csv",parse_dates=["Date"],index_col="Date")
df.head(5)
```

```python
# Ensure the index is a DatetimeIndex
df.index = pd.to_datetime(df.index)
# print(df.index)
# Sort the index if it's not already sorted
df = df.sort_index()

# Now you can slice the DataFrame by date range
df["2017-01-01":"2017-01-07"]
```

```python
%matplotlib inline
df.Close.resample("w").mean().plot(kind="bar")
```
