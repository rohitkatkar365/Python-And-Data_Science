```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("weather_data1.csv")
df
```

```python
df1 = df.replace(-99999,np.nan)
df1
```

```python
df.at[2, "windspeed"] = -88888
df
```

```python
# If you want to replace multiple value
df2 = df.replace([-99999,-88888],np.nan)
df2
```

```python
# Another way
df3 = df.replace({
    "temperature":-99999,
    "windspeed":[-99999,-88888],
    "event":"0"
},np.nan)
df3
```

```python
df4 = df.replace({"-99999":np.nan,
                  -88888:np.nan,
                 '0':"No Event"})
df4
```
