```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv",parse_dates=["day"])
type(df.day[0]) # Return string
# Convert into Data Type parse_dates=["day"]
df.set_index("day",inplace=True)
df
```

```python
# How to handle missing data (Use dataframe_name.fillna(val))
df1 = df.fillna(0)
df1
```

```python
# If you want to add value to the particular column values
df2 = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event': 'No Event'
})
df2
```

```python
# 0.0 temperature no mean instead we use forward fillb
df3 = df.fillna(method="ffill")
df3
```

```python
df4 = df.fillna(method="bfill")
df4
```

```python
df5 = df.fillna(method="ffill",limit=1)
df5
```

```python
# The interpolate() method is powerful for filling in gaps in your data, and you can choose the interpolation method that best suits your dataset. 
df6 = df.interpolate()
df6
```

```python
# Remove missing value row.records
df7 = df.dropna()
df7
```

```python
df8 = df.dropna(how="all")
df8
```

```python
df9 = df.dropna(thresh=1)
print(df)
df9
```
