```python
import pandas as pd
```

```python
df = pd.read_excel("sd.xlsx")
df
```

```python
def convert_people_name(cell):
    if cell == 'n.a.':
        return "RRK"
    return cell

df2 = pd.read_excel("sd.xlsx",converters={
    "people":convert_people_name
})
df2
```

```python
df3 = pd.read_excel("sd.xlsx",na_values=["not available","n.a."])
df3
```

```python
def convert_revenue(cell):
    if cell == -1:
        return None
    return cell

df2 = pd.read_excel("sd.xlsx",converters={
    "revenue":convert_revenue
})
df2
```
