```python
# If header is missing in csv file
df1 = pd.read_csv("stock_data.csv",header=None,names=["ticker","eps","revene","price","people"])
df1
```

```python
# If i want read/fetch only 5 or 10 record from large dataset.
# Use nrows argument
df2 = pd.read_csv("stock_data.csv",header=None,names=["ticker","eps","revene","price","people"],nrows=3)
df2
```

```python
# How to turn not available value to NaN
df3 = pd.read_csv("stock_data.csv",na_values=["not available","n.a."])
df3
```

```python
# Another way (This is best way)
df4 = pd.read_csv("stock_data.csv",na_values={
'eps':["not available","n.a."],
'revenue': ["not available","n.a.",-1]
,'price':["not available","n.a."],
    'people':["not available","n.a."]
})
df4
```

```python
# How to write in csv
df4.to_csv("new.csv",index=False,columns=["tickers","eps"]) 
# By using index argument : remove index
# By Using Columns argument : Write only particular column to csv file
```
